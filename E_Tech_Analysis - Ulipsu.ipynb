```python
# loading the Dataset for Analysis.

df = pd.read_csv("/content/IV Assignment Data - Usage Details.csv")
df1 = pd.read_csv(("/content/IV Assignment Data - Enrollment Details.csv"))
df2 = pd.read_csv("/content/IV Assignment Data - Course_theme_details.csv")
```

```python
# Removing Sec from the Duration
df['Duration_without_sec'] = df['Duration'].apply(lambda x: x if 'sec' not in x else x.replace('sec', '').strip())

# Display the DataFrame
print(df)
```

```python
df
```

```python
# converting Sec into HH:MM:SS format

# Function to convert duration to "hh:mm:ss"
def convert_to_hms(duration):
    if 'sec' in duration:
        seconds = int(duration.replace('sec', '').strip())
    else:
        time_parts = list(map(int, duration.split(':')))
        seconds = time_parts[0] * 3600 + time_parts[1] * 60 + time_parts[2]
    return f'{seconds // 3600:02d}:{(seconds % 3600) // 60:02d}:{seconds % 60:02d}'

# Apply the function to the Duration column and replace the existing column
df['Duration'] = df['Duration'].apply(convert_to_hms)

# Display the DataFrame
print(df)
```

```python
df.columns = df.columns.str.strip()
```

```python
# convert Yes or No into Numerical Value
df['Assessment status'] = df['Assessment status'].replace('YES', 1)

# If you want to convert the column to numeric type after replacement
df['Assessment status'] = pd.to_numeric(df['Assessment status'], errors='coerce')
```

```python
# convert Yes or No into Numerical Value
df['Project Status'] = df['Project Status'].replace('YES', 1)
df['Project Status'] = df['Project Status'].replace('NO', 0)

# If you want to convert the column to numeric type after replacement
df['Project Status'] = pd.to_numeric(df['Project Status'], errors='coerce')
```

```python
# Visualize the distribution of categorical variables
cat_cols = ['Region', 'Cluster', 'Implementation Mode', 'Category Of Schools',
            'Assessment status', 'Project Status']  # Add other categorical columns as needed

for col in cat_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Correlation matrix for numerical columns
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Each Field Analysis Plot

# 1. School Code, Region, Cluster, Implementation Mode, Category Of Schools
categorical_columns = ['School Code', 'Region', 'Cluster', 'Implementation Mode', 'Category Of Schools']

for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.show()

# 2. Course Id
plt.figure(figsize=(20, 6))
sns.countplot(x='Course Id', data=df)
plt.title('Distribution of Course Id')
plt.show()

# 3. Course Watched Percentage
plt.figure(figsize=(20, 6))
sns.histplot(df['Course Watched Percentage'], kde=True)
plt.title('Distribution of Course Watched Percentage')
plt.show()

# 4. Assessment status
plt.figure(figsize=(8, 5))
sns.countplot(x='Assessment status', data=df)
plt.title('Distribution of Assessment status')
plt.show()

# 5. Project Status
plt.figure(figsize=(8, 5))
sns.countplot(x='Project Status', data=df)
plt.title('Distribution of Project Status')
plt.show()


# 6. Weight
plt.figure(figsize=(10, 6))
sns.histplot(df['weight'], kde=True)
plt.title('Distribution of Weight')
plt.show()
```
